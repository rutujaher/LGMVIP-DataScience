# Task 2:Image to pencil sketch with python

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img=cv2.imread('Lion.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("Original Image")
plt.imshow(img)
```

```python
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(imggray,cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

```python
imginvert=cv2.bitwise_not(imggray)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(imginvert,cmap="gray")
plt.title("Inverted Image")
plt.show()
```

```python
imgsmooth=cv2.GaussianBlur(imginvert, (21,21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(imgsmooth,cmap="gray")
plt.title("Smoothen Image")
plt.show()
```

```python
imgsketch=cv2.divide(imggray, 255-imgsmooth, scale=255)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(imgsketch,cmap="gray")
plt.title("Final Pencil Sketch")
plt.show()
```
